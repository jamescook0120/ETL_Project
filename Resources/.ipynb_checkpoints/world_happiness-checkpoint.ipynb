```python
rds_connection_string = "postgres:postgres@localhost:5432/happiness_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
world_happiness_report_df.to_sql(name='world_happiness_2008_2019', con=engine, if_exists='append', index=False)
```

```python
world_happiness_report_2021_df.to_sql(name='world_happiness_2021', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from world_happiness_2008_2019', con=engine).head()
```
